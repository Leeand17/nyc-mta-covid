```python
# iterating the columns 
ny_zip = df[df["STATE"] == "NY"]
print(ny_zip)
```

```python
ny_zipcodes = ny_zip.loc[:, ['zipcode','agi_stub', 'N1','A00100']]
print(ny_zipcodes)
print(ny_zipcodes.dtypes)
```

```python
def calcAGI(d): 
    # {1:1000, 2: 1000, }
    total = sum(d.values())
    return 1*d[1]/total + 2*d[2]/total + 3*d[3]/total + 4*d[4]/total + 5*d[5]/total + 6*d[6]/total
```

```python
prev_zip = 0
zip_val = {}
data = []
for index, row in ny_zipcodes.iterrows():
    if prev_zip == row['zipcode']:
        zip_val[row['agi_stub']] = row['N1']
    else: 
        agi_avg = calcAGI(zip_val)
        data.append([prev_zip, agi_avg])
        prev_zip = row['zipcode']
        zip_val[row['agi_stub']] = row['N1']
zip_df = pd.DataFrame(data, columns = ['ZIPCODE', 'agi_avg']) 
zip_df = zip_df.astype({"ZIPCODE": int})
zip_df
```

```python
# Merge on zipcode
data = zip_df.merge(ny_turnstile, on = 'ZIPCODE', how = 'inner')
print(data)
print(data.keys())
```

```python
data = data.loc[:, ['ZIPCODE','agi_avg']]
print(data)
```
